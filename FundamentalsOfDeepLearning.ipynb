**Training a** **Perceptron**

```python
import numpy as np

class Perceptron:

   def __init__(self, learning_rate = 0.01, num_iterations = 1000):
      self.lr = learning_rate
      self.n_iters = num_iterations
      # Changed to call the internal _relu_function. Also, the class method name should be consistent (self.activation_function = self._relu_function)
      self.activation_function = self._relu_function
      self.weights = None
      self.bias = None

   def fit(self, X, y):
      n_Samples, n_features = X.shape

      # Initialize weights and biases

      self.weights = np.zeros(n_features)
      self.bias = 0

      for _ in range(self.n_iters):
        for idx, x_i in enumerate(X):
          linear_output = np.dot(x_i, self.weights) + self.bias
          y_predicted = self.activation_function(linear_output)

          # Perceptron Update Rule
          # Fixed typo: y_[idx] should be y[idx]
          update = self.lr * (y[idx] - y_predicted)
          self.weights += update * x_i
          self.bias += update

   # Corrected indentation for _relu_function and predict methods
   # Also, `Self` should be `self` as a convention for instance methods
   def _relu_function(self, x):
      return np.maximum(0, x)

   def predict(self, X):
      linear_output = np.dot(X, self.weights) + self.bias
      y_predicted = self.activation_function(linear_output)
      return y_predicted
```

```python
X = np.array([[0,0], [0,1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])
```

```python
perceptron = Perceptron(learning_rate = 0.1, num_iterations = 10)
perceptron.fit(X, y)
```

```python
print("Predictions: ", perceptron.predict(X))
```

**Activation Functions -** **Demo** **bold text**

```python
import tensorflow as tf
```

```python
from tensorflow import keras
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = tf.linspace(-10, 10, 100)
X
```

```python
X.shape
```

```python
# Sigmoid Activation
# Activations
y = tf.keras.activations.sigmoid(X)

# Visualize Graph
plt.plot(X, y, label = "Sigmoid = 1 / (1 + exp(-x))")
plt.title("Sigmoid")
plt.xlabel("X")
plt.ylabel("Sigmoid of X")
plt.legend()
plt.show()
```

```python
# tan h

#Activations
y = tf.keras.activations.tanh(X)

#Visualize Graph
plt.plot(X, y, label = "tanh(X)")
plt.title("Hyperbolic Tangent")
plt.xlabel("X")
plt.ylabel("tanh(X)")
plt.legend()
plt.show()
```

```python
# Relu

#Activations
y = tf.keras.activations.relu(X)

#Visualize Graph
plt.plot(X, y, label = "y = max(x, 0)")
plt.title("Relu")
plt.xlabel("X")
plt.ylabel("max(x, 0)")
plt.legend()
plt.show()
```

```python
# Leaky Relu

#Activations
y = tf.nn.leaky_relu(X)

#Visualize Graph
plt.plot(X, y)
plt.title("Leaky Relu")
plt.xlabel("X")
plt.ylabel("Leaky Relu")
plt.legend()
plt.show()
```

```python
# Elu
# Activations
y = tf.keras.activations.elu(X, alpha = 0.9)

#Visualize Graph
plt.plot(X, y, label = "y = x if x>0 else alpha * (exp(x) - 1) if x < 0")
plt.title("Elu")
plt.xlabel("X")
plt.ylabel("Elu(X)")
plt.legend()
plt.show()
```

```python
#Softmax
x = tf.constant([150, 50, 10], dtype = tf.float32)
x = tf.expand_dims(x, axis = 0)
print(x.shape, x)
y = tf.keras.activations.softmax(x)
print(y)
```
