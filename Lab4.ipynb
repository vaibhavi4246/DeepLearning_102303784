# **Multiple Linear Regression using Linear Perceptron**

**Step 1 :  Load and Inspect Dataset**

```python
import pandas as pd

data = pd.read_csv("/content/multiple_linear_regression_dataset.csv")

print(data.head())
print(data.columns)
print(data.shape)
```

**Step 2 : Separate Inputs and Output**

```python
X = data[['age', 'experience']].values
y = data['income'].values
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

**Step 3 : Initialize Parameters**

```python
import numpy as np

n_features = X.shape[1]

w = np.zeros(n_features)
b = 0.0

print("Initial weights:", w)
print("Initial bias:", b)
```

**Step 4 : Forward Pass**

```python
def predict(X, w, b):
    y_hat = X.dot(w) + b
    return y_hat
```

**Step 5 : Mean Squared Error**

```python
def mean_squared_error(y, y_hat):
    loss = ((y_hat - y) ** 2).mean()
    return loss
```

**Step 6 : Compute Gradients**

```python
def compute_gradients(X, y, y_hat):
    N = len(y)

    dw = (2 / N) * X.T.dot(y_hat - y)
    db = (2 / N) * (y_hat - y).sum()

    return dw, db
```

**Step 7 : Update Parameters**

```python
def update_parameters(w, b, dw, db, lr):
    w = w - lr * dw
    b = b - lr * db
    return w, b
```

**Step 8 : Training Loop**

```python
lr = 0.0001
epochs = 1000

for epoch in range(epochs):

    y_hat = predict(X, w, b)

    loss = mean_squared_error(y, y_hat)

    dw, db = compute_gradients(X, y, y_hat)

    w, b = update_parameters(w, b, dw, db, lr)

    if epoch % 100 == 0:
        print("Epoch:", epoch, "Loss:", loss)

print("Final weights:", w)
print("Final bias:", b)
```

**Step 9 : Final Evaluation**

```python
new_candidate = np.array([4.5, 68])

predicted_salary = new_candidate.dot(w) + b

print("Predicted Salary:", predicted_salary)
```

**Effect of Learning Rate**

If learning rate is large:

*  Loss increases
*  Training unstable

If learning rate is small:

Training very slow

*   Training very slow

So we choose small value like 0.0001 because salary values are large.
