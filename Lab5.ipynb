```python
import numpy as np
import pandas as pd
```

```python
column_names = [
    "Sex", "Length", "Diameter", "Height",
    "Whole_weight", "Shucked_weight",
    "Viscera_weight", "Shell_weight", "Rings"
]

df = pd.read_csv("abalone.data", header=None, names=column_names)
print(df.head())
```

```python
df["y"]=df["Rings"]+1.5
```

```python
features = ["Length", "Diameter", "Whole_weight"]
X=df[features].values
y=df["y"].values.reshape(-1, 1)
```

```python
N = len(X)
split=int(0.8*N)

X_train = X[:split]
X_test = X[split:]
y_train = y[:split]
y_test = y[split:]

print("Train shape:", X_train.shape, y_train.shape)
print("Test shape:", X_test.shape, y_test.shape)
```

```python
mean=X_train.mean(axis=0)
std=X_train.std(axis=0)

X_train=(X_train-mean)/std
X_test=(X_test-mean)/std
```

```python
def forward(X, w, b):
    """
    Computes y_hat=Xw + b
    """
    y_hat=np.dot(X, w) + b

    print("Shape of X:", X.shape)
    print("Shape of w:", w.shape)
    print("Shape of b:", np.shape(b))
    print("Shape of y_hat:", y_hat.shape)

    return y_hat
```

```python
def mse(y, y_hat):
    return np.mean((y-y_hat)**2)
```

```python
def grad_w(X, y, y_hat):
    N=len(y)
    return (-2/N)*np.dot(X.T, (y-y_hat))

def grad_b(y, y_hat):
    N=len(y)
    return (-2/N)*np.sum(y-y_hat)
```

```python
np.random.seed(42)

d=X_train.shape[1]

w=np.random.randn(d, 1)*0.01
b=0.0

learning_rate=0.01
epochs=200

for epoch in range(epochs):
    y_hat=np.dot(X_train, w)+b
    loss=mse(y_train, y_hat)
    dW=grad_w(X_train, y_train, y_hat)
    db=grad_b(y_train, y_hat)

    assert dW.shape==w.shape

    w=w-learning_rate*dW
    b=b-learning_rate*db

    if epoch%20==0:
        print("Epoch:", epoch, "Loss:", loss)
```

```python
y_test_hat=np.dot(X_test, w) + b

test_mse=mse(y_test, y_test_hat)
test_mae=np.mean(np.abs(y_test - y_test_hat))

print("\nTest MSE:", test_mse)
print("Test MAE:", test_mae)

print("\nFive sample predictions:\n")
for i in range(5):
    true_val=y_test[i][0]
    pred_val=y_test_hat[i][0]
    abs_error=abs(true_val-pred_val)

    print(f"True: {true_val:.2f} | Predicted: {pred_val:.2f} | Abs Error: {abs_error:.2f}")
```
